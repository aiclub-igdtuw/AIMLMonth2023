## Question-3: Can you identify any significant predictors of body fat percentage? Which body measurements have the strongest correlation with body fat percentage?

```python
import pandas as pd

# Read the dataset
data = pd.read_csv('bodyfat.csv')

# Task 3: Identify significant predictors of body fat percentage
correlation = data.corr()['BodyFat'].sort_values(ascending=False)
significant_predictors = correlation[abs(correlation) > 0.5].index.tolist()

print("Significant predictors of body fat percentage:")
print(significant_predictors)

# Strongest correlation with body fat percentage
strongest_correlation = correlation[0]

print("Strongest correlation with body fat percentage:")
print(strongest_correlation)
```

## Analysis
Conducted correlation analysis to identify significant predictors of body fat percentage.
Determined body measurements with the strongest correlation to body fat percentage.
## Observation
Body fat percentage has a strong positive correlation with weight and abdomen circumference.
Other body measurements, such as height and neck circumference, have weaker correlations with body fat percentage.

## Question-5: Visualize the relationship between body fat percentage and each of the body measurements using scatter plots or line plots with multiple axes

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
data = pd.read_csv('bodyfat.csv')

# Task 5: Visualize the relationship between body fat percentage and body measurements
measurements = ['Age', 'Weight', 'Height', 'Neck', 'Chest', 'Abdomen', 'Hip', 'Thigh',
                'Knee', 'Ankle', 'Biceps', 'Forearm', 'Wrist']

# Create a figure with multiple subplots
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(12, 12))

# Iterate over each body measurement and plot the relationship with body fat percentage
for i, measurement in enumerate(measurements):
    row = i // 4
    col = i % 4
    ax = axes[row][col]

    ax.scatter(data[measurement], data['BodyFat'], alpha=0.5)
    ax.set_xlabel(measurement)
    ax.set_ylabel('Body Fat Percentage')
    ax.set_title(f'{measurement} vs Body Fat Percentage')

# Adjust the spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```

## Analysis
Visualized the relationship between body fat percentage and each body measurement using scatter plots or line plots with multiple axes.
Examined the patterns and trends in theÂ relationships.
## Observation
Scatter plots show a positive linear relationship between body fat percentage and weight, as well as abdomen circumference.
Other body measurements, such as height and neck circumference, do not exhibit a clear linear relationship with body fat percentage.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

# Data processing
# Read the dataset
df = pd.read_csv("bodyfat.csv")
df.head(10)  # <-- Added parentheses after df.head

# Further data processing steps and analysis can be performed here
```

```python

X = df.drop(['BodyFat','Density'],axis=1)
y = df['Density']

X['Bmi']=703*X['Weight']/(X['Height']*X['Height'])
X.head()
```

```python
df['NavyFat'] = 495 / (1.0324 - 0.19077 * np.log10(df['Abdomen'] - df['Neck']) + 0.15456 * np.log10(df['Height'] * 100)) - 450
df['NavyFat'] = round(df['NavyFat'], 1)
```

```python
df[['BodyFat', 'NavyFat', 'Abdomen']][df['BodyFat'] < 10]
```

```python
df.columns
```

```python
X = df[['Age', 'Weight', 'Height', 'Neck', 'Chest',
       'Abdomen', 'Hip', 'Thigh', 'Knee', 'Ankle', 'Biceps', 'Forearm',
       'Wrist']]
y_b = df['BodyFat']
y_n = df['NavyFat']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train_b, y_test_b, y_train_n, y_test_n = train_test_split(X, y_b,y_n,
                                                                             test_size=0.2,
                                                                            random_state=42)
```

```python
print(X_train.shape, X_test.shape)
print(y_train_b.shape, y_test_b.shape)
print(y_train_n.shape, y_test_n.shape)
```

```python
from sklearn.metrics import mean_absolute_error
model = LinearRegression()

model.fit(X_train, y_train_b) #training the model for BodyFat column

y_pred_b = model.predict(X_test)
mae_b = mean_absolute_error(y_test_b, y_pred_b)

print(f'the mean absolute error for linear regression trained on BodyFat column is {round(mae_b, 2)}')

model.fit(X_train, y_train_n) #re-training the model for NavyFat column

y_pred_n = model.predict(X_test)
mae_n = mean_absolute_error(y_test_n, y_pred_n)

print(f'the mean absolute error for linear regression trained on NavyFat column is {round(mae_n, 2)}')
```

# Model prediction

```python
final_graph = X_test.copy()
final_graph['BodyFat'] = y_test_b.copy()
final_graph['NavyFat'] = y_test_n.copy()
final_graph['y_pred_b'] = y_pred_b.copy()
final_graph['y_pred_n'] = y_pred_n.copy()

final_graph.head()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))

sns.scatterplot(ax=ax[0], data=final_graph, x='Abdomen', y='BodyFat')
sns.scatterplot(ax=ax[0], data=final_graph, x='Abdomen', y='y_pred_b')

sns.scatterplot(ax=ax[1], data=final_graph, x='Abdomen', y='NavyFat')
sns.scatterplot(ax=ax[1], data=final_graph, x='Abdomen', y='y_pred_n')

plt.show()
```
