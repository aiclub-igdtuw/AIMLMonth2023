<-----------------------Question 3----------------------------->

Print the list of events with year in the end.

```python
# Extract the year from the event date
data['year'] = pd.to_datetime(data['published_date']).dt.year

# Create a new column with the event name and year
data['event_with_year'] = data['event'] + ' (' + data['year'].astype(str) + ')'

# Print the list of events with the year
event_list_with_year = data['event_with_year'].unique()
for event in event_list_with_year:
    print(event)
```

<-----------------------Question 4----------------------------->

Plot a histogram 'Published Date vs talks published that month

```python
# Extract the month from the 'published_date' column
data['published_month']=pd.to_datetime(data['published_date']).dt.month

# Count the number of talks published in each month
talks_per_month = data['published_month'].value_counts().sort_index()

# Create a histogram of talks published per month
plt.bar(talks_per_month.index, talks_per_month.values)

plt.xlabel('Published Month')
plt.ylabel('Number of Talks Published')
plt.title('Number of Talks Published per Month')

# Customize the x-axis tick labels to display month names instead of numbers
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(talks_per_month.index, month_names)
```

Summary/Explanation of the above code :

the code counts the number of talks published in each month using .value_counts().sort_index(), which sorts the counts based on the month order. A histogram is created using plt.bar() with the months on the x-axis and the number of talks published on the y-axis. Labels, title, and custom x-axis tick labels (month names) are added to the plot.

**DATA PRE-PROCESSING**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
talks = pd.read_csv("talks_info.csv")
```

```python
import re
# Function to convert numeric values with suffixes to integers
def convert_numeric(value):
    suffixes = {'K': 1000, 'M': 1000000}
    match = re.search(r'^(\d*\.?\d*)([KM])$', value)
    if match:
        number = float(match.group(1))
        suffix = match.group(2)
        return int(number * suffixes[suffix])
    else:
        return int(value)

# Apply the conversion function to the 'likes' column
talks['likes'] = talks['likes'].apply(convert_numeric)
```

```python
talks['likes']
```

```python
x = talks[['duration','views']]
y =talks ['likes']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test ,y_train ,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model =LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
c=model.intercept_
```

```python
c
```

```python
m=model.coef_
m
```

```python
y_pred_train=model.predict(x_train)
```

```python
plt.scatter(y_train,y_pred_train)
plt.xlabel("Actual ")
plt.ylabel("predicted")
plt.show()
```
