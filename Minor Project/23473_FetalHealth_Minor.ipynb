#Fetal Health

### What is CTG?

Cardiotocography (CTG) is employed throughout pregnancy for the purpose of monitoring the fetal heart rate and uterine contractions. Primarily utilized in the third trimester, its main objective is to ensure the well-being of the fetus and enable the early identification of any signs of fetal distress. Should an abnormal CTG reading be observed, it may necessitate additional investigations and potential intervention.

###How does CTG works?
The device utilized in cardiotocography is commonly referred to as a cardiotocograph. It entails the placement of two transducers on the abdomen of a pregnant woman. One transducer utilizes ultrasound to record the fetal heart rate, while the other transducer monitors uterine contractions by measuring the tension of the maternal abdominal wall, offering an indirect indication of intrauterine pressure. Following this, the cardiotocograph (CTG) is evaluated by a midwife and the obstetric medical team.

##**Q4. Can we create visualizations to understand the relationship between different CTG features (e.g., acceleration, deceleration) and the fetal health states?**
