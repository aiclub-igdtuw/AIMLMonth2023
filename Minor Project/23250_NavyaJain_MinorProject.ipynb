####Analyze the distribution and range of values for each clinical and demographic feature in the dataset. Are there any outliers or extreme values?

```python
features = ['Age', 'Sex', 'ALB', 'ALP', 'ALT', 'AST', 'BIL', 'CHE', 'CHOL', 'CREA', 'GGT', 'PROT']
```

```python
features_stats = data[features].describe()
print(features_stats)
```

```python
for features in features:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[features], kde=True)
    plt.title(f'Distribution of {features}')
    plt.xlabel(features)
    plt.ylabel('Count')
    plt.show()
```

```python

# Select the numerical columns
numerical_columns = ['Age', 'ALB', 'ALP', 'ALT', 'AST', 'BIL', 'CHE', 'CHOL', 'CREA', 'GGT', 'PROT']

# Identify and print the outliers for each numerical column
for column in numerical_columns:
    # Calculate the box plot statistics
    stats = data[column].describe()
    q1 = stats['25%']
    q3 = stats['75%']
    iqr = q3 - q1

    # Define the outlier threshold
    lower_threshold = q1 - 10 * iqr
    upper_threshold = q3 + 10 * iqr

    # Identify the outliers
    outliers = data[(data[column] < lower_threshold) | (data[column] > upper_threshold)][column]

    # Print the outliers
    if len(outliers) > 0:
        print(f"Outliers in {column}:")
        print(outliers)
        print()
    else:
        print(f"No outliers found in {column}")
        print()
```

Summarizing your analysis and observation

```python
#Outliers identified with a threshhold multiplier of 10 since we are finding only extreme values
```

####Can you explore the correlations between features and the diagnosis of hepatitis C? Are there any strong correlations or dependencies?

```python
corr_matrix = data.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()

# Identify strong correlations
threshold = 0.5
strong_correlations = corr_matrix[(corr_matrix > threshold) | (corr_matrix < -threshold)]
strong_correlations = strong_correlations.dropna(how='all', axis=0).dropna(how='all', axis=1)

# Print the strong correlations
print("Strong Correlations:")
print(strong_correlations)
```
