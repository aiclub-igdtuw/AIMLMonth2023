<h3> Q4. How can you predict the luckiest number for future Powerball drawings? Are there any machine learning algorithms or statistical techniques that can be employed for this prediction task?

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Select the ball number columns
numbers = df[['w1', 'w2', 'w3', 'w4', 'w5', 'w6']]

# Split the data into training and testing sets
train_size = int(len(numbers) * 0.8)
train_data, test_data = numbers[:train_size], numbers[train_size:]

# Fit SARIMAX models and make predictions
predictions = []
rmse_values = []
for column in train_data.columns:
    model = SARIMAX(train_data[column], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    model_fit = model.fit()
    preds = model_fit.predict(start=train_size, end=len(numbers) - 1)
    predictions.append(preds)
    
    # Calculate RMSE for each feature
    mse = mean_squared_error(test_data[column], preds)
    rmse = mse ** 0.5
    rmse_values.append(rmse)

# Print individual RMSE values
for i, column in enumerate(train_data.columns):
    print(f"RMSE for {column}: {rmse_values[i]}")






```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load the data into a DataFrame
# Assuming 'df' contains the Powerball dataset with columns: Draw Date, w1, w2, w3, w4, w5, w6

# Select the ball number columns
numbers = df[['w1', 'w2', 'w3', 'w4', 'w5', 'w6']]

# Split the data into training and testing sets
train_size = int(len(numbers) * 0.8)
train_data, test_data = numbers[:train_size], numbers[train_size:]

# Fit ARIMAX models and make predictions
predictions = []
rmse_values = []
exog_train = train_data.drop(columns='w1')  # Use all other columns as exogenous variables

for column in train_data.columns:
    exog_test = test_data.drop(columns=column)
    
    model = ARIMA(train_data[column], order=(1, 1, 1), exog=exog_train)
    model_fit = model.fit()
    preds = model_fit.predict(start=train_size, end=len(numbers) - 1, exog=exog_test)
    predictions.append(preds)
    
    # Calculate RMSE for each feature
    mse = mean_squared_error(test_data[column], preds)
    rmse = mse ** 0.5
    rmse_values.append(rmse)

# Print individual RMSE values
for i, column in enumerate(train_data.columns):
    print(f"RMSE for {column}: {rmse_values[i]}")
```

```python
# Add more cells if required
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error

# Load the data into a DataFrame
# Assuming 'df' contains the Powerball dataset with columns: Draw Date, w1, w2, w3, w4, w5, w6

# Select the ball number columns to predict
target_columns = ['w1', 'w2', 'w3', 'w4', 'w5', 'w6']
target_data = df[target_columns].values

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(target_data)

# Define the number of time steps to consider
time_steps = 7

# Create sequences and labels
sequences = []
labels = []
for i in range(time_steps, len(scaled_data)):
    sequences.append(scaled_data[i - time_steps: i])
    labels.append(scaled_data[i])
sequences = np.array(sequences)
labels = np.array(labels)

# Split the data into training and testing sets
train_size = int(len(sequences) * 0.8)
train_sequences, test_sequences = sequences[:train_size], sequences[train_size:]
train_labels, test_labels = labels[:train_size], labels[train_size:]

# Reshape the data for LSTM input (samples, time steps, features)
train_sequences = np.reshape(train_sequences, (train_sequences.shape[0], time_steps, len(target_columns)))
test_sequences = np.reshape(test_sequences, (test_sequences.shape[0], time_steps, len(target_columns)))

# Define and train the LSTM model
model = Sequential()
model.add(LSTM(64, input_shape=(time_steps, len(target_columns))))
model.add(Dense(len(target_columns)))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(train_sequences, train_labels, epochs=50, batch_size=32)

# Make predictions on the test set
predictions = model.predict(test_sequences)

# Inverse scale the predictions and labels
predictions = scaler.inverse_transform(predictions)
test_labels = scaler.inverse_transform(test_labels)

# Calculate RMSE for each target feature
rmse = []
for i in range(len(target_columns)):
    mse = mean_squared_error(test_labels[:, i], predictions[:, i])
    rmse.append(np.sqrt(mse))

print("Root Mean Squared Error (RMSE) for each target feature:")
for i in range(len(target_columns)):
    print(f"{target_columns[i]}: {rmse[i]}")
```
