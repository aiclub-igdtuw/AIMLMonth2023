**Q5) Can you analyze the pairwise correlations between the features? Are there any features that are highly correlated with each other? How might you handle multicollinearity if present?**

```python
corr= df.corr()
print(corr)
```

```python
high_corr= df.drop("species", axis=1).corr()
high_corr= corr.abs() > 0.8
corr_features = set()
for i in range(len(high_corr.columns)):
    for j in range(i):
        if high_corr.iloc[i, j]:
            colname = high_corr.columns[i]
            corr_features.add(colname)

print("Highly correlated features:", corr_features)
```

```python
sns.heatmap(data=df.corr(), annot=True, cmap="coolwarm")
plt.title('Pairwise Correlation between the features')
plt.show()
```


1) **Sepal length** has a strong positive correlation with **petal length (0.87)** and **petal width (0.82)**

2) **Sepal width** has a weak negative correlation with **sepal length (-0.11)** and **petal width (-0.36)** and **petal length (-0.42)**

3) **Petal length** and **petal width** have the strongest positive correlation **(0.96)** This high correlation can lead to multicollinearity. To handle this we can either remove one of the highly correlated features or use techniques like Ridge regression that reduces correlated features and stabilises the coefficients.

Summarize your model's performance by evaluation metrices

```python
from sklearn.model_selection import train_test_split
X = df.drop('species', axis=1)
Y = df['species']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred, average='weighted')
print('Accuracy:', accuracy)
print('Precision:', precision)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(X_train, Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred, average='weighted')
print('Accuracy:', accuracy)
print('Precision:', precision)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred, average='weighted')
print('Accuracy:', accuracy)
print('Precision:', precision)
```

Among Logistic Regression, KNeighbours Classifier and Decision Tree Classifier, the highest accuracy we got is **96.67%**  and precision 96.92% is from **KNeighbours Classifier**.
