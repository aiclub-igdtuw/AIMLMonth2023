# Observation:
-> Outliers are present in our dataset and thus comes the need to remove them

```python
# Removal of outliers

def mod_outlier(wine1):
        df = wine1.copy()
        wine1 = wine1._get_numeric_data()
        q1 = wine1.quantile(0.25)
        q3 = wine1.quantile(0.75)

        iqr = q3 - q1

        lower_bound = q1 -(1.5 * iqr) 
        upper_bound = q3 +(1.5 * iqr)


        for col in wine1.columns:
            for i in range(0,len(wine1[col])):
                if wine1[col][i] < lower_bound[col]:            
                    wine1[col][i] = lower_bound[col]
                    
                if wine1[col][i] > upper_bound[col]:            
                    wine1[col][i] = upper_bound[col]    


        for col in wine1.columns:
            df[col] = df[col]

        return(df)

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,cross_validate
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
X = wine1.drop('quality', axis=1)
y = wine1['quality']
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE(k_neighbors=4)
# transform the dataset
X, y = oversample.fit_resample(X, y)
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
def classify(model, X, y):
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    # train the model
    model.fit(x_train, y_train)
    print("Accuracy:", model.score(x_test, y_test) * 100)
    
    # cross-validation
    score = cross_val_score(model, X, y, cv=5)
    print("CV Score:", np.mean(score)*100)
```

```python
model = DecisionTreeClassifier()
classify(model, X, y)
```

```python
model = RandomForestClassifier()
classify(model, X, y)
```
