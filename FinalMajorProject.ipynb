````python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
import librosa
import librosa.display
from IPython.display import Audio
import warnings
warnings.filterwarnings('ignore')
````

````python
def waveplot(data, sr, emotion):
    plt.figure(figsize=(10,4))
    plt.title(emotion, size=20)
    librosa.display.waveplot(data, sr=sr)
    plt.show()

def spectogram(data, sr, emotion):
    x = librosa.stft(data)
    xdb = librosa.amplitude_to_db(abs(x))
    plt.figure(figsize=(11,4))
    plt.title(emotion, size=20)
    librosa.display.specshow(xdb, sr=sr, x_axis='time', y_axis='hz')
    plt.colorbar()
````

###Feature Extraction

````python
def extract_mfcc(filename):
    y, sr = librosa.load(filename, duration=3, offset=0.5)
    mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)
    return mfcc
````

````python
extract_mfcc(df['speech'][0])
````

````python
X_mfcc = df['speech'].apply(lambda x: extract_mfcc(x))
````

````python
X_mfcc
````

````python
X= [x for x in X_mfcc]
X= np.array(X)
X.shape
````

````python
#split input
X= np.expand_dims(X,-1)
X.shape
````

````python
from sklearn.preprocessing import OneHotEncoder
enc= OneHotEncoder()
y= enc.fit_transform(df[['label']])
````

````python
y= y.toarray()
````

````python
y.shape
````

###Creating the Model (LSTM)

````python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout

model= Sequential([
    LSTM(256, return_sequences= False, input_shape=(40,1)),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation= 'relu'),
    Dropout(0.2),
    Dense(7, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'] )
model.summary()
````

````python
#training the model
history= model.fit(X,y, validation_split=0.2, epochs=50, batch_size=64)
````

Best value accuracy: 78.93

````python
#Stacked LSTM
model_stacked = Sequential([
    LSTM(256, return_sequences=True, input_shape=(40, 1)),
    LSTM(128, return_sequences=False),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(7, activation='softmax')
])
model_stacked.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_stacked.summary()
````

````python
#Transformer
from keras.layers import Input, Dense, MultiHeadAttention, Dropout
from keras.models import Model

inputs = Input(shape=(40, 1))
x = MultiHeadAttention(key_dim=256, num_heads=4)(inputs, inputs)
x = Dropout(0.2)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.2)(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.2)(x)
outputs = Dense(7, activation='softmax')(x)

model_transformer = Model(inputs, outputs)
model_transformer.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_transformer.summary()
````
