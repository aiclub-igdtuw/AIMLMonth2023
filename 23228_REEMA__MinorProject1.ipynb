```python
# Define a function to extract image properties

def extract_image_properties(image_path):
    image = Image.open(image_path)
    width, height = image.size
    mode = image.mode
    file_format = os.path.splitext(image_path)[1].replace(".", "")
    image_array = np.array(image)
    image_quality = np.mean(image_array)

    return {
        'Image Path': image_path,
        'Resolution': f'{width}x{height}',
        'Color/Grayscale': mode,
        'File Format': file_format,
        'Image Quality': image_quality
    }
    # Define the dataset folder path

    # Replace with the actual path to the dataset folder

dataset_path = '/content/drive/MyDrive/archive'

# Create an empty list to store image properties

image_properties = []

# Iterate through the dataset and extract image properties

for root, _, files in os.walk(dataset_path):
    for file in files:
        if file.endswith(('.jpg', '.png')):
            image_path = os.path.join(root, file)
            properties = extract_image_properties(image_path)
            image_properties.append(properties)

# Create a pandas DataFrame from the image properties list

df = pd.DataFrame(image_properties)

# Display the image properties

print(df.head())
```

```python
import csv

def read_csv_file(file_path):
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        rows = []
        for row in reader:
            rows.append(row)
    return rows

# Provide the path to your CSV file

csv_file_path = '/content/drive/MyDrive/archive/Mechanical_Tools_labelled.csv'
data = read_csv_file(csv_file_path)

# You can now work with the 'data' variable, which contains the contents of the CSV file as a list of rows.

# Each row is represented as a list of values.

# Example: Print the contents of the CSV file
for row in data:
    print(row)
```

```python
import csv

def convert_csv_to_text(csv_file_path, text_file_path):
    with open(csv_file_path, 'r') as csv_file:
        reader = csv.reader(csv_file)
        with open(text_file_path, 'w') as text_file:
            for row in reader:

                # Modify this logic to extract the desired content from the CSV file
                # Use '\t' as a delimiter for the text file

                line = '\t'.join(row)
                text_file.write(line + '\n')

# Provide the path to your CSV file

csv_file_path = '/content/drive/MyDrive/archive/Annotated.csv'

# Provide the desired path for the text file

text_file_path = '/content/abc.txt'

convert_csv_to_text(csv_file_path, text_file_path)
```
