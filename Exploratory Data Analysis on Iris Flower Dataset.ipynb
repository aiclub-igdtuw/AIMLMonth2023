<-----------------------Question 5----------------------------->

Can you analyze the pairwise correlations between the features? Are there any features that are highly correlated with each other? How might you handle multicollinearity if present?

```python
#----------code
```

```python
sns.pairplot(iris)
```

```python
#pairwise correlation
iris_corr = iris.corr()
sns.heatmap(iris_corr, annot=True, cmap='coolwarm',linewidths=0.4)
plt.title("Correlation between features using Heatmap")
fig=plt.gcf()
fig.set_size_inches(8,6)
plt.show()
```

```python
corr_matrix = iris.corr().abs()

print(corr_matrix)
```

```python
correlation_matrix = iris.drop("SPECIES NAME", axis=1).corr()

#finding highly correlated features
high_correlation = correlation_matrix.abs() > 0.8
correlated_features = set()
for i in range(len(high_correlation.columns)):
    for j in range(i):
        if high_correlation.iloc[i, j]:
            colname = high_correlation.columns[i]
            correlated_features.add(colname)

print("Highly correlated features:", correlated_features)
```

**Analysis and Observation**

*The high correlation coefficients observed between certain features indicate a strong linear relationship among them. In this plot, Petal Length and Petal Width exhibit a very high positive correlation of 0.96.*
*Multicollinearity is when 3 or more features are highly correlated with each other. In this dataset, the characteristics (sepal length, sepal width, petal length, and petal width) are not highly associated.*

*To remove multicollinearity, we can:*

*1) Remove one of the features from the highly correlated pairs.*

*2) Add the independent variables together or else combine them linearly or alter them to produce new features with lower correlation or more independence.*

*3)Multicollinearity is handled by regularisation techniques like Lasso or Ridge regression, which reduce correlated features and stabilise coefficients.*
