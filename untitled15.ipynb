```python
data=data.drop(columns=['rest_type', 'dish_liked', 'url', 'address', 'phone', 'cuisines'])
```

```python
yes_no_map={'Yes':1 , 'No': 0}
```

```python
data['online_order']=data['online_order'].map(yes_no_map)
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
# Group the data by the 'online_order' column and count the number of restaurants in each group
online_order_counts = data.groupby('online_order')['name'].count()

# Define the x-axis labels and positions
x_labels = [0,1]
x_pos = [0, 1]

# Create the bar plot
plt.bar(x_pos, online_order_counts, align='center')

# Set the x-axis labels and tick positions
plt.xticks(x_pos, x_labels)

# Set the y-axis label
plt.ylabel('Number of Restaurants')

# Set the plot title
plt.title('Online Ordering vs. Non-Online Ordering Restaurants')

# Show the plot
plt.show()
```

```python
data['rate'].replace({"/5":'','NEW|-':'0'},regex=True,inplace=True)
data['rate']= data['rate'].astype('float')
data['rate'].fillna(data['rate'].mean(), inplace = True)
data['rate'].isnull().sum()
```

```python
data
```

```python
# Group the data by the 'approx_cost_for_two_people' column and calculate the mean rating for each group
cost_rating = data.groupby('approx_cost(for two people)')['rate'].mean().reset_index()

# Set the figure size and plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(cost_rating['approx_cost(for two people)'], cost_rating['rate'])

# Set the x-axis and y-axis labels and title
plt.xlabel('Approx Cost for Two People')
plt.ylabel('Rating')
plt.title('Average Rating by Approx Cost for Two People')
plt.

# Show the plot
plt.show()

```
